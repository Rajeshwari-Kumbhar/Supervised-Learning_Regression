```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
# Splitting dataset into Independent and Dependent variable

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#Fitting Random forest regression model

from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=10, random_state=0)
reg.fit(x,y)
```

```python
# Predict for myinput
reg.predict([[4]])
```

```python
# Visualization

from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0],filled=True)
plt.title("Tree plot")
plt.show()
```

```python
# Predict for user input

level=float(input("Enter the level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)

print(f"Predicted salary for level {level}: ₹{result[0]}")
```
