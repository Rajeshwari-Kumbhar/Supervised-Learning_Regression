```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing and reading dataset

ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Splitting dataset into Independent and Dependent data

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Plot graph                (Relationship between two variables) 

plt.scatter(x,y)
plt.title("Scatter plot")
plt.show()
```

```python
# Model fitting (Polynomial regression model fitting)

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x)
#x_poly
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
# Visualization

plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='red')
plt.title("Scatter plot")
plt.show()
```

```python
# Predicting for my input

myinput=pd.DataFrame(data=[[11]],columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result
```

```python
# Predict for user input

level=float(input("Enter Level:"))
myinput=pd.DataFrame(data=[[level]],columns=['Level'])
poly_input=poly_reg.transform(myinput)
result=reg.predict(poly_input)                  
print(f"Predicted result for Level {level}: {round(result[0,0])}")                  
```
