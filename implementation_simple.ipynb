```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset

ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
ds.columns
```

```python
# Divide dataset as Independent (input) and Dependent(Output) variables

x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
# Scatter plot graph                               (Positive correlation)

plt.scatter(x,y)
plt.title('Scatter plot')
plt.show()
```

# If experience year increase Salary will increase

```python
x
```

```python
y
```

```python
# Splitting dataset as training(80%) and testing(20%)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Introduce Model              (fitting model)

from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Predict

result=reg.predict([[15]])
result
```

```python
result=reg.predict(x_test)
result
```

```python
y_test   # Cross checking
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
# Checking accuracy on y_train and y_test

from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```

```python
# Visualize Plot 

plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='red') # Color of line can be changed
plt.title("Scatter plot")
plt.show()
```

```python
# For user input

years=int(input("Enter yeras of experience:"))
myinput=pd.DataFrame(columns=['YearsExperience'],data=[[years]])
result=reg.predict(myinput)
print("Your salary must close to",round(result[0,0]))
```
