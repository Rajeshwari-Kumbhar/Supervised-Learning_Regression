```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset

ds=pd.read_csv("startup_data.csv")
ds
```

```python
ds.columns
```

```python
# Divide dataset as Independent and Dependent variables

x=ds[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y=ds[['Profit']]
```

```python
x
```

```python
y
```

```python
# Finding unique values in State column

ds['State'].unique()
```

```python
# Encoding on column state
# OHE done by alphabetically
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data=ohe.fit_transform(x[['State']])
```

```python
encoded_data
```

```python
column_names = ohe.get_feature_names_out(['State'])
encoded_df = pd.DataFrame(encoded_data,columns = column_names)
```

```python
encoded_df
```

```python
x= pd.concat([x,encoded_df],axis=1)
x
```

```python
x = x.drop(columns=['State'])
x
```

```python
x
```

```python
# Splitting dataset as training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
# Fit Model  (Multiple linear Regression Model)

from sklearn.linear_model import  LinearRegression
reg =  LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Predict

data = [[100000,100000,100000,1,0,0]]
columns= x.columns
myinput= pd.DataFrame(columns = columns, data = data)
result= reg.predict(myinput)
result
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Checking accuracy on y_train and y_test

from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```

```python
# Visualization

plt.scatter(x["R&D Spend"],y["Profit"])
plt.scatter(x["Administration"],y["Profit"],color='orange')
plt.scatter(x["Marketing Spend"],y["Profit"],color='green')
plt.title("Scatter plot")
plt.show()
```
